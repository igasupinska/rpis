```python
import numpy as np
import matplotlib.pyplot as plt

def poissonDistribution(l, n):
    randoms = np.random.poisson(l, n)
    plt.hist(randoms, 50, density=True)
    plt.show()
    print('Mean of Poisson distributions is', sum(randoms)/len(randoms))
    print('Standard deviation of Poisson distribution is', np.std(randoms))
    
    averages = [sum(randoms[0:i])/(i+1) for i in range(n)]
    lambdaLine = np.array([sum(randoms)/len(randoms) for i in range(n)])
    stdDev = np.subtract(lambdaLine, averages)
    plt.plot(averages)
    plt.plot(stdDev)
    plt.show()
    
poissonDistribution(10, 1000)
#TO DO extra task
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

def discreteDistWithoutMean():
    probabilities = {}
    for i in range(-10, 10, 1):
        if i != 0:
            probabilities[i] = 1/(4*abs(i)*(abs(i)+1))
        else:
            probabilities[i] = 0.5

    lists = sorted(probabilities.items())
    x, y = zip(*lists)
    plt.plot(x, y)
    plt.show()
    
    #sampling
    randoms = [probabilities[random.choice(list(probabilities))] for i in range (100)]
    plt.plot(randoms)
    plt.show()
    
discreteDistWithoutMean()
```
